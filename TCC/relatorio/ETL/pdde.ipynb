```python
import pandas as pd
import re
```

```python
# Função para remover tags HTML de uma célula
def remove_html_tags(cell):
    if isinstance(cell, str):  # Verifica se a célula é uma string
        return re.sub(r'<[^>]*>', '', cell).strip()  # Remove as tags HTML
    return cell
```

```python
# Parâmetros
bloco_tamanho = 20  # 18 linhas de dados + 2 linhas de separação
dados_por_bloco = 18  # Quantidade de linhas úteis por bloco
```

```python
# Carregar o arquivo
file_path = "C:/Users/User/Downloads/TCC_DADOS/pdde_2019.xlsx"  
df = pd.read_excel(file_path, header=None)
```

```python
output_file = "C:/Users/User/Downloads/TCC_DADOS/trsn/pdde_2019.csv" 
```

```python
with open(output_file, mode='w', encoding='utf-8') as f:
    for i in range(0, len(df), bloco_tamanho):
        # Seleciona as 18 linhas úteis do bloco
        bloco = df.iloc[i:i+dados_por_bloco]
        
        # Verifica se o bloco tem a quantidade correta de linhas úteis
        if len(bloco) == dados_por_bloco:
            bloco = bloco.applymap(remove_html_tags)  # Remove as tags HTML
            bloco = bloco.T  # Transpõe o bloco
            bloco.to_csv(f, header=False, index=False, lineterminator='\n')  # Salva o bloco no arquivo
        else:
            print(f"O bloco {i//bloco_tamanho} não possui a quantidade correta de linhas úteis")
```
