## Question 00

```python
# installing empiricaldist
# installed it before on the enviroment for Midterm

import numpy as np
import empiricaldist as emp
import pandas as pd
import seaborn as sns
```

## Question 01

```python
# 01.01: Load the "brfss.hdf5" dataset into a dataframe
df = pd.read_hdf('brfss.hdf5', 'brfss')
df.head()
```

```python
# 01.02: Extract the variable 'AGE' from the DataFrame brfss and assign it to age.
# 01.03: Plot the PMF of age as a bar chart.

def answer_MidTermExam_Part02_one():
    age = df['AGE']
    pmf_age = emp.Pmf.from_seq(age)
    # making the bars larger
    pmf_age.bar(width=2)
    # adding labels
    plt = emp.plt
    plt.xlabel('Age')
    plt.ylabel('PMF')
    return plt

answer_MidTermExam_Part02_one()
```

## Question 02

```python
# 02.01: Make a scatter plot of weight and age with format string 'o' and alpha=0.1. Hint: Use plt.plot() rather than plt.scatter.

def answer_MidTermExam_Part02_two():
    age = df['AGE']
    plt = emp.plt
    weight = df['WTKG3']
    plt.plot(age, weight, 'o', alpha=0.1)
    plt.xlabel('Age in years')
    plt.ylabel('Weight in kg')
    return plt

answer_MidTermExam_Part02_two()
```

## Question 03

```python
# 03.01: Add random noise to age with mean 0 and standard deviation 2.5. Hint: Use np.random.normal() jitter age.
# 03.02: Make a scatter plot between weight and age with marker size 5 and alpha=0.2. 
# Be sure to also specify 'o'. Hint: Use plt.plot() rather than plt.scatter. 
# Make sure that age is the first argument passed. The parameter for marker size is markersize.

def answer_MidTermExam_Part02_three():
    age = df['AGE'] + np.random.normal(0, 2.5, size=len(df))
    plt = emp.plt
    weight = df['WTKG3']
    plt.plot(age, weight, 'o', markersize=5, alpha=0.2)
    plt.xlabel('Age in years')
    plt.ylabel('Weight in kg')
    return plt

answer_MidTermExam_Part02_three()
```

## Question 04

```python
# 04.01: Fill in the parameters of .boxplot() to plot the distribution of weight ('WTKG3') in each height ('_HTMG10') group. 
# Specify whis=10, just as was done in the video. Hint: The parameters you should set are x, y, data, and whis.
# 04.02: Add a line to plot the y-axis on a logarithmic scale. Hint: Use the plt.yscale() function to specify a logarithmic scale.

def answer_MidTermExam_Part02_four():
    plt = emp.plt
    # dropping NA values
    df4 = df.dropna(subset=['_HTMG10', 'WTKG3']) 
    sns.boxplot(x='_HTMG10', y='WTKG3', data=df4, whis=10)
    plt.yscale('log')
    plt.xticks(rotation=90)
    plt.xlabel('Height in cm')
    plt.ylabel('Weight in kg')
    return plt

answer_MidTermExam_Part02_four()
```

## Question 05

```python
# 05.01: Extract 'INCOME2' from the brfss DataFrame and assign it to income.  
# 05.02: Plot the PMF of income as a bar chart. Hint: Use Pmf() to create a PMF and .bar() to make a bar chart.

def answer_MidTermExam_Part02_five():
    income = df['INCOME2']
    pmf_income = emp.Pmf.from_seq(income)
    pmf_income.bar()
    plt = emp.plt
    plt.xlabel('Income in $')
    plt.ylabel('PMF')
    return plt

answer_MidTermExam_Part02_five()
```

## Question 06

```python
# 06.01: Create a violin plot to plot the distribution of height ('HTM4') in each income ('INCOME2') group. Specify inner=None to simplify the plot. 
# Hint: Using sns.violinplot(), the parameters you should set are x, y, data, and inner. Each income group should be on the x-axis.

def answer_MidTermExam_Part02_six():
    plt = emp.plt
    df6 = df.dropna(subset=['HTM4', 'INCOME2'])
    sns.violinplot(x='INCOME2', y='HTM4', data=df6, inner=None)
    plt.ylabel('Height in cm')
    plt.xlabel('Income in $')
    return plt

answer_MidTermExam_Part02_six()
```

## Question 07

```python
# 07.01: From the brfss DataFrame, select the columns 'AGE', 'INCOME2', and '_VEGESU1'.
# 07.02: Compute the correlation matrix for these variables. Hint: After selecting the columns, use the .corr() method to compute the correlation matrix.

def answer_MidTermExam_Part02_seven():
    df7 = df[['AGE', 'INCOME2', '_VEGESU1']]
    corr = df7.corr()
    return corr

answer_MidTermExam_Part02_seven()
```

## Question 08

**Which of the following are correct interpretations of these results:**

A: People with higher incomes eat more vegetables.

B: The relationship between income and vegetable consumption is linear.

C: Older people eat more vegetables.

D: There could be a strong nonlinear relationship between age and vegetable consumption.

**Possible Answers:**

(I) A and C only

(II) B and D only

(III) B and C only

(IV) A and D only

**My Answer:** 

(II) B and D only

After checking the barplots, scatterplots and correlation scores, the consumption of vegetables is not strongly associated with any of the variables. But it is possible to see a small relation with income. When we analyze the plots, we can also see that from 2-8 income units, the consumption of vegetables grows larger, which could give the idea of a linear relationship. 

```python
# checking linear relationship between income and vegetable consumption (barplot)

def answer_MidTermExam_Part02_eight_AB():
    plt = emp.plt
    df8 = df.dropna(subset=['INCOME2', '_VEGESU1'])
    sns.barplot(x='INCOME2', y='_VEGESU1', data=df8)
    plt.xlabel('Income in $')
    plt.ylabel('Vegetable consumption')
    return plt

answer_MidTermExam_Part02_eight_AB()
```

```python
# checking linear relationship between income and vegetable consumption (scatterplot)

def answer_MidTermExam_Part02_eight_AB():
    plt = emp.plt
    df8 = df.dropna(subset=['INCOME2', '_VEGESU1'])
    sns.scatterplot(x='INCOME2', y='_VEGESU1', data=df8)
    plt.xlabel('Income in $')
    plt.ylabel('Vegetable consumption')
    return plt

answer_MidTermExam_Part02_eight_AB()
```

```python
# checking if older people consume more vegetables (barplot)

def answer_MidTermExam_Part02_eight_C():
    plt = emp.plt
    df8 = df.dropna(subset=['AGE', '_VEGESU1'])
    sns.barplot(x='AGE', y='_VEGESU1', data=df8)
    plt.xlabel('Age in years')
    plt.ylabel('Vegetable consumption')
    return plt

answer_MidTermExam_Part02_eight_C()
```

```python
# checking if older people consume more vegetables (scatterplot)

def answer_MidTermExam_Part02_eight_C():
    plt = emp.plt
    df8 = df.dropna(subset=['AGE', '_VEGESU1'])
    sns.scatterplot(x='AGE', y='_VEGESU1', data=df8)
    plt.xlabel('Age in years')
    plt.ylabel('Vegetable consumption')
    return plt

answer_MidTermExam_Part02_eight_C()
```
