Now, the second aproach is using geopandas and geoplot to plot the grades of each state. We can use the dataset of the grades and the dataset of the geometry of each state to plot the grades of each state on the reference data (Link 3).

```python
# create a map using geoplotlib for "MEDIA_EM_MT" column
import geopandas as gpd
import geoplot as gplt
import geoplot.crs as gcrs
import matplotlib.pyplot as plt
import pandas as pd

# load the map of Brazil
brasil = gpd.read_file('BR_UF_2022/BR_UF_2022.shp')

# rename CD_UF to ID_UF
brasil.rename(columns={'CD_UF':'ID_UF'}, inplace=True)
# make sure its the same type
brasil['ID_UF'] = brasil['ID_UF'].astype(int)
brasil.head()

# Assuming df_base is already defined and contains the 'ID_UF' column
# merge the two datasets
df = pd.merge(df_base, brasil, on='ID_UF')

# Convert the merged DataFrame to a GeoDataFrame
df = gpd.GeoDataFrame(df, geometry='geometry')

# Ensure that the geometries are valid
df['geometry'] = df['geometry'].apply(lambda geom: geom if geom.is_valid else geom.buffer(0))

# plot the map
fig, ax = plt.subplots(1, 1, figsize=(15, 15))
gplt.choropleth(df, hue='MEDIA_EM_MT', cmap='viridis', linewidth=0.5, ax=ax, legend=True)
plt.title('Average Math Score by State')
plt.show()
```
