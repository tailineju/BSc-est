```python
import pandas as pd
```

```python
dataset = pd.read_csv('path/cars.csv')
dataset.head()
```

```python
#Q1.
dataset.isnull().sum()
```

```python
import numpy as np
num_dataset = dataset.select_dtypes(np.number)
for col in num_dataset.columns:
    col_mean = num_dataset[col].mean()
    num_dataset[col] = num_dataset[col].fillna(col_mean)

print("\nDataset with null values filled:\n", num_dataset.head())
```
