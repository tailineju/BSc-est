```python
import pandas as pd
```

```python
dataset = pd.read_csv('path/police.csv')
dataset
```

```python
#Q1.
null_counts = dataset.isnull().sum()

# Drop columns where null count matches number of rows
new_dataset = dataset.drop(null_counts[null_counts == len(dataset)].index, axis=1)
```

```python
new_dataset
```

```python
#Q2.
Gender = dataset['driver_gender']
Gender
```

```python
Gender_values = Gender.value_counts()
```

```python
Gender_values
```

```python
Gender_values['M'] > Gender_values['F']
```

```python
print('were Men or Women stopped more often:', 'Men')
```

```python
#Q3.
gender_dataset = new_dataset.groupby('driver_gender')['search_conducted']
gender_dataset.sum()
```

```python
gender_dataset.sum()['F'] < gender_dataset.sum()['M']
```

```python
print('Does gender affect who gets searched during a stop?', 'Yes')
```

```python
#Q4.
mapping_dataset = new_dataset['stop_duration'].map({'0-15 Min': 7.5, '16-30 Min': 23, '30+ Min': 45}).mean()
```
