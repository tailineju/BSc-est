```python
# importing needed library
import pandas as pd
```

```python
# loading the dataset
dataset = pd.read_csv('C:/Users/User/Documents/GitHub/gradest-1/DATAVIS/tutorials/Tutorial04/Datasets/covid19.csv')
```

```python
# checking dataset's information
dataset.info()
```

```python
# Q. 1) Show the number of Confirmed, Deaths and Recovered cases in each Region.
dataset.groupby('Region')[['Confirmed', 'Deaths', 'Recovered']].sum()
```

```python
# Q. 2) Remove all the records where the Confirmed Cases is Less Than 10.
dataset = dataset[dataset['Confirmed'] >= 10]
```

```python
# Q. 3) In which Region, maximum number of Confirmed cases were recorded ?
dataset[dataset['Confirmed'] == dataset['Confirmed'].max()]['Region']
```

```python
# Q. 4) In which Region, minimum number of Deaths cases were recorded ?
# number of rows with min deaths
dataset[dataset['Deaths'] == dataset['Deaths'].min()].shape[0]
```

```python
# Q. 4) In which Region, minimum number of Deaths cases were recorded ?
dataset[dataset['Deaths'] == dataset['Deaths'].min()][['Region', 'Deaths']].head()
```

```python
# Q. 5) How many Confirmed, Deaths & Recovered cases were reported from India till 29 April 2020 ?
dataset[dataset['Region'] == 'India'][['Confirmed', 'Deaths', 'Recovered']]
```

```python
# Q. 6-A ) Sort the entire data wrt No. of Confirmed cases in ascending order.
dataset.sort_values(by='Confirmed', ascending=True)
```

```python
# Q. 6-B ) Sort the entire data wrt No. of Recovered cases in descending order.
dataset.sort_values(by='Recovered', ascending=False)
```
