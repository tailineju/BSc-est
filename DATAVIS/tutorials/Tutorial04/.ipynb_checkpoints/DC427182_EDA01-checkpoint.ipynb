```python
# importing needed library
import pandas as pd
```

```python
# loading the dataset
dataset = pd.read_csv('C:/Users/User/Documents/GitHub/gradest-1/DATAVIS/tutorials/Tutorial04/Datasets/weather.csv')
```

```python
# checking the basic information of the data
dataset.info()
```

By the previous output, we can say that the dataset has 7 columns (variables) and 8784 rows (observations). 

```python
# checking the first 5 rows of the data
dataset.head()
```

```python
# Q.1-1) finding the unique values of Wind Speed
dataset["Wind Speed_km/h"].unique()
```

```python
# Q.1-2) number of times when the Weather is exactly 'Clear'
dataset[dataset["Weather"] == "Clear"].shape[0]
```

```python
# Q.1-3) number of times when the Wind Speed was exactly '4 km/h'
dataset[dataset["Wind Speed_km/h"] == 4].shape[0]
```

```python
# Q.1-4) Find out all the Null Values in the data
dataset.isnull().sum()
```

```python
# Q.1-5) renaming the column name 'Weather' of the dataframe to 'Weather Condition'
dataset.rename(columns={'Weather': 'Weather Condition'}, inplace=True)
dataset.columns
```

```python
# Q.1-6) calculating the mean for 'Visibility'
dataset["Visibility_km"].mean()
```

```python
# Q.1-7) calculating the Standard Deviation of 'Pressure'
dataset['Press_kPa'].std()
```

```python
# Q.1-8) calculating the Variance of 'Relative Humidity'
dataset['Rel Hum_%'].var()
```

```python
# Q.1-9) finding all instances when 'Snow' was recorded.
dataset[dataset['Weather Condition'].str.contains('Snow', na=False)]
```

```python
# Q.1-10) finding all instances when 'Wind Speed is above 24' and 'Visibility is 25'.
dataset[(dataset['Wind Speed_km/h'] > 24) & (dataset['Visibility_km'] == 25)]
```

```python
# Q.1-11)calculating the mean of each column by 'Weather Condition'?
dataset.groupby('Weather Condition').mean(numeric_only=True).head()
```

```python
# Q.1-12) finding the Minimum value of each column by 'Weather Condition'
dataset.groupby('Weather Condition').min(numeric_only=True).head()
```

```python
# Q.1-12) finding the Maximum value of each column by 'Weather Condition'
dataset.groupby('Weather Condition').max(numeric_only=True).head()
```

```python
# Q.1-13) showing all the Records where Weather Condition is Fog.
dataset[dataset['Weather Condition'] == 'Fog']
```

```python
# Q.1-14) finding all instances when 'Weather is Clear' or 'Visibility is above 40'.
dataset[(dataset['Weather Condition'] == 'Clear') | (dataset['Visibility_km'] > 40)]
```

```python
# Q.1-15) Find all instances when : A. 'Weather is Clear' and 'Relative Humidity is greater than 50' or B. 'Visibility is above 40'
dataset[((dataset['Weather Condition'] == 'Clear') & (dataset['Rel Hum_%'] > 50)) | (dataset['Visibility_km'] > 40)]
```
