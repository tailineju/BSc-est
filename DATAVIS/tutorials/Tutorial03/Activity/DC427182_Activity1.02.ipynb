**Derive insights from the sizes of forest fires**

```python
# 04: filter entries with area larger than 0
f_dataset=dataset[dataset['area'] > 0]
f_dataset
```

```python
# 05: calculating mean, min, max, and std of the area
print("Mean=",f_dataset['area'].mean())
print("Min=",f_dataset['area'].min())
print("Max=",f_dataset['area'].max())
print("Std=",f_dataset['area'].std())
```

```python
# 06: sorting the filtered dataset using the area
f_dataset.sort_values('area').tail(20)
```

```python
# 07: calculating the median of the area and visually compare it to the mean value
print("Median=",f_dataset['area'].median())
print("Mean=",f_dataset['area'].mean())
print("The mean could be influenced by big values that are not representative for the whole dataset.")
```

**Finding the month with the most forest fires**

```python
# 08: getting a list of unique values from the month column
f_dataset['month'].unique()
```

```python
# 09: getting the number of entries for the month of March using shape 
f_dataset[f_dataset['month'] == 'mar'].shape[0]
```

```python
# 10: iterating over all the months, filter our dataset for the rows containing the given month, and calculate the mean temperature.
# print a statement with the number of fires, the mean temperature, and the month.
for month in f_dataset['month'].unique():
    month_data = f_dataset[f_dataset['month'] == month]
    print(f"{month} \n fires: {month_data.shape[0]} \n mean temperature: {month_data['temp'].mean()}")
```
