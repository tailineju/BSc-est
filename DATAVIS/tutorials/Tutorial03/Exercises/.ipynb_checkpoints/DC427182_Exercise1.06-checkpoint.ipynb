## Indexing

```python
# 02: importing pandas with an alias
import pandas as pd
```

```python
# 03: loading the dataset
dataset = pd.read_csv('C:/Users/User/Documents/GitHub/gradest-1/DATAVIS/tutorials/Tutorial03/Datasets/world_population.csv', index_col=0)
```

```python
# 04: locating a row by value
dataset.loc[["United States"]].head()
```

```python
# 05: locating a row by index
dataset.iloc[[-2]]
```

```python
# 06: indexing only the values from column 2000
dataset["2000"].head()
```

```python
# 07: finding India on the indexed dataset (column 2000)
dataset[["2000"]].loc[["India"]]
```

```python
# 08: the same as in step 7, but now having a raw number output
dataset["2000"].loc["India"]
```

## Slicing

```python
# 09: slicing countries of rows 2 to 5
dataset.iloc[1:5]
```

```python
# 10: slicing Germany, Singapore, United States, and India
dataset.loc[["Germany", "Singapore", "United States", "India"]]
```

```python
#11: slicing a list of countries and their population density for specific years (columns)
country_list = ["Germany", "Singapore", "United States", "India"]
dataset.loc[country_list][["1970", "1990", "2010"]]
```

## Iterating

```python
# 12: iterating all countries until Angola (excluded)
for index, row in dataset.iterrows():
    if index == 'Angola':
        break
    print(index, '\n', \
          row[["Country Code", "1970", "1990", "2010"]], '\n')
```
