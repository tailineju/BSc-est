```python
# 01: importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# 02: reading the data
dataset = pd.read_csv('C:/Users/User/Documents/GitHub/gradest-1/DATAVIS/tutorials/Tutorial05/Datasets/smartphone_sales.csv', index_col=1)
dataset.drop('Unnamed: 0', axis=1, inplace=True)
dataset.head()
```

```python
# 03: creating a stacked area chart
plt.figure(figsize=(10, 6))
dataset.plot.area(stacked=True)
plt.title('Smartphone Brand Sales')
plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.legend(loc='upper right')
plt.show()
```
