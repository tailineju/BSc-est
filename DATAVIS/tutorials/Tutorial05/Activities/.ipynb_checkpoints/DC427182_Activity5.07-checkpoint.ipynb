```python
# 01: importing libraries
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

```python
# 02:  Load all four images from the Datasets subfolder: namely, "images.zip"

base_dir = r'C:\Users\User\Documents\GitHub\gradest-1\DATAVIS\tutorials\Tutorial05\Datasets\Images'

# List of image files
image_files = [os.path.join(base_dir, f'photo-{i}.jpg') for i in range(1, 5)]

# Load images
images = []
for img_file in image_files:
    if os.path.exists(img_file):
        images.append(mpimg.imread(img_file))
    else:
        print(f"File not found: {img_file}")

# Verify loaded images
for i, img in enumerate(images):
    print(f"Image {i+1} loaded with shape: {img.shape}")
```

```python
# 03: Display the four images in a 2x2 grid using matplotlib
titles = ['Coast', 'Beach', 'Building', 'City at Night']

fig, axs = plt.subplots(2, 2, figsize=(10, 10))
for i, ax in enumerate(axs.flat):
        ax.imshow(images[i])
        ax.axis('off')
        ax.set_title(titles[i])

plt.show()
```
