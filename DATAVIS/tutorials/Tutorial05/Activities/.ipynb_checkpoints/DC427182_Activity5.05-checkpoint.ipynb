```python
# 03: ploting a histogram
# Plot a histogram with 10 bins for the given IQ scores. IQ scores are normally distributed with a mean of 100 and a standard deviation of 15. Visualize the mean as a vertical solid red line, and the standard deviation using dashed vertical lines. Add labels and a title.
plt.hist(iq_scores, bins=10, color='skyblue', edgecolor='black')
plt.axvline(np.mean(iq_scores), color='red', linestyle='solid', linewidth=2)
plt.axvline(np.mean(iq_scores) + np.std(iq_scores), color='red', linestyle='dashed', linewidth=2)
plt.axvline(np.mean(iq_scores) - np.std(iq_scores), color='red', linestyle='dashed', linewidth=2)
plt.xlabel('IQ Scores')
plt.ylabel('Frequency')
plt.title('IQ Scores Distribution')
plt.show()
```
