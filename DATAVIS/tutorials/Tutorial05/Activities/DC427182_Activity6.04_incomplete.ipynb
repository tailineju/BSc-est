```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# 01: loading the data
dataset = pd.read_csv('C:/Users/User/Documents/GitHub/gradest-1/DATAVIS/tutorials/Tutorial05/Datasets/YouTube.csv')
#rename the columns
dataset.columns = ['Channel','Subs','Views']
# subs: float; views: float
dataset['Subs'] = dataset['Subs'].astype(float)
dataset['Views'] = dataset['Views'].astype(float)
dataset.head()
```

```python
# 02-03: reformating the data
dataset=pd.melt(dataset, id_vars=['Channel'], value_vars=['Subs','Views'], var_name='Metric', value_name='Count')
dataset.head()


```

```python
# 04: Create a FacetGrid with two columns to visualize the data.
# y: channel
# x: count
# facet: metric

plt.figure(figsize=(12,6))
sns.set(style='whitegrid')
plt.subplot(1,2,1)
sns.barplot(x='Subs', y='Channel', data=dataset, color='blue')
plt.title('Barplot')
plt.subplot(1,2,2)
sns.boxplot(x='Views', y='Channel', data=dataset, color='orange')
plt.title('Boxplot')

plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12, 6))
sns.set(style='whitegrid')

# Create a GridSpec for the subplots
from matplotlib.gridspec import GridSpec
gs = GridSpec(1, 2, width_ratios=[1, 1], wspace=0.05)

# Barplot
ax1 = plt.subplot(gs[0])
sns.barplot(x='Subs', y='Channel', data=dataset, color='blue', ax=ax1)
ax1.set_title('Barplot')
ax1.set_xlabel('Subs')
ax1.set_ylabel('Channel')

# Boxplot
ax2 = plt.subplot(gs[1], sharey=ax1)
sns.boxplot(x='Views', y='Channel', data=dataset, color='orange', ax=ax2)
ax2.set_title('Boxplot')
ax2.set_xlabel('Views')
ax2.set_ylabel('')

# Remove the y-axis labels from the second plot
ax2.yaxis.set_tick_params(labelleft=False)

plt.tight_layout()
plt.show()
```
