```python
# 01: importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 02: loading the data
dataset = pd.read_csv('C:/Users/User/Documents/GitHub/gradest-1/DATAVIS/tutorials/Tutorial05/Datasets/anage_data.csv', index_col=0)
```

```python
# 03: Filter the data so that you end up with samples containing a body mass and maximum longevity as the given dataset, AnAge, which was used in the previous exercise, is not complete. 
# Select all of the samples of the Aves class with a body mass of less than 20,000. 
dataset = dataset[dataset['Class'] == 'Aves']
dataset = dataset[dataset['Body mass (g)'] < 20000]
dataset.head()
```

```python
# 04: Create a Figure with a constrained layout. Create a gridspec of size 4x4. Create a scatter plot of size 3x3 and marginal histograms of size 1x3 and 3x1. Add labels and a Figure title.

fig = plt.figure(constrained_layout=True)

gs = fig.add_gridspec(4, 4)
ax = fig.add_subplot(gs[1:, :-1])
ax_hist_x = fig.add_subplot(gs[0, :-1], sharex=ax)
ax_hist_y = fig.add_subplot(gs[1:, -1], sharey=ax)

ax.scatter(dataset['Body mass (g)'], dataset['Maximum longevity (yrs)'])
ax.set_xlabel('Body mass (g)')
ax.set_ylabel('Maximum longevity (yrs)')
ax_hist_x.hist(dataset['Body mass (g)'], bins=30, orientation='vertical')
ax_hist_y.hist(dataset['Maximum longevity (yrs)'], bins=30, orientation='horizontal')
```
