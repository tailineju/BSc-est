```python
# 02: importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.ticker import FuncFormatter
```

```python
# 03: loading the dataset
dataset = pd.read_csv('C:/Users/User/Documents/GitHub/gradest-1/DATAVIS/tutorials/Tutorial05/Datasets/movie_scores.csv', index_col=0)
```

```python
# 03: checking the dataset
# renaming the columns
dataset.columns = ['Movie', 'Tomatometer', 'Audience Score']
dataset.head()
```

```python
# 04--06: creating a visually appealing bar plot comparing the two scores for all five movies.

# fixing one of the movie's name to fit the plot better
dataset.loc[dataset['Movie'] == 'The Hobbit: An Unexpected Journey', 'Movie'] = 'The Hobbit:\nAn Unexpected Journey'

# defining the format for the y-axis
def percent_formatter(x, pos):
    return f'{int(x)}%'

# creating the bar plot
bar_width = 0.35

r1 = np.arange(len(dataset['Movie']))
r2 = [x + bar_width for x in r1]

plt.figure(figsize=(10, 6))

ax = plt.gca()
ax.set_xticks([r + bar_width / 2 for r in r1])
ax.set_xticklabels(dataset['Movie'])

ax.set_yticks(range(0, 101, 20))
ax.set_yticks(range(0, 101, 5), minor=True)
ax.yaxis.grid(True, which='major', zorder=0)
ax.yaxis.grid(True, which='minor', linestyle='--', zorder=0)
ax.yaxis.set_major_formatter(FuncFormatter(percent_formatter))

plt.bar(r1, dataset['Tomatometer'], color='blue', width=bar_width,
        label='Tomatometer', zorder=3)
plt.bar(r2, dataset['Audience Score'], color='orange', width=bar_width,
        label='Audience Score', zorder=3)

plt.xlabel('Movie')
plt.ylabel('Scores')
plt.title('Movie Scores')
plt.legend()

plt.show()
```
