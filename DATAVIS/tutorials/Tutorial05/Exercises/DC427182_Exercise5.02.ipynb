```python
# 02: import libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 03: loading the dataset
dataset = pd.read_csv('C:/Users/User/Documents/GitHub/gradest-1/DATAVIS/tutorials/Tutorial05/Datasets/water_usage.csv', index_col=0)
```

```python
# 04:  pie chart to visualize water usage

# Create figure
plt.figure(figsize=(8, 8), dpi=300)
# Create pie plot
plt.pie('Percentage', explode=(0, 0, 0.1, 0, 0, 0), \
        labels='Usage', data=dataset, autopct='%.0f%%')
# Add title
plt.title('Water usage')
# Show plot
plt.show()
```
