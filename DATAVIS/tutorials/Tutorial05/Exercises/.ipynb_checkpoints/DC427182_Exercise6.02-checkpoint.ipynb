```python
# 02: import libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# 03: load data
data = pd.read_csv("C:/Users/User/Documents/GitHub/gradest-1/DATAVIS/tutorials/Tutorial05/Datasets/northern_surface_temperature.csv", index_col=['Year'])
data = data.transpose()
```

```python
# 04:  creating color palette
heat_colormap = sns.diverging_palette(240, 15, s=99,as_cmap=True)
```

```python
# 05: plot heatmap
plt.figure(dpi=200)
sns.heatmap(data.iloc[:, ::5], cmap=heat_colormap, center=0)
plt.title("Temperature Changes from 1880 to 2015 " \
          "(base period 1951-1980)")
plt.savefig('img/temperature_change.png', dpi=300, \
            bbox_inches='tight')
```
