```python
# 02: importing libraries
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 03: loading the dataset
mydata = pd.read_csv('C:/Users/User/Documents/GitHub/gradest-1/DATAVIS/tutorials/Tutorial05/Datasets/iq_scores.csv')
```

```python
# 04: creating test groups
group_a = mydata[mydata.columns[0]].tolist()
group_b = mydata[mydata.columns[1]].tolist()
group_c = mydata[mydata.columns[2]].tolist()
group_d = mydata[mydata.columns[3]].tolist()
```

```python
# 05: checking the groups
print(group_a)
print(group_b)
print(group_c)
print(group_d)
```

```python
# 06: creating the dataframe
data = pd.DataFrame({'Groups': ['Group A'] \
                     * len(group_a) + ['Group B'] \
                     * len(group_b) + ['Group C'] \
                     * len(group_c) + ['Group D'] \
                     * len(group_d),\
                     'IQ score': group_a + group_b \
                     + group_c + group_d})
```

```python
# 07: checking the dataframe
print(data.columns)
```

```python
# 08: creating the boxplot
plt.figure(dpi=150)
# Set style
sns.set_style('whitegrid')
# Create boxplot
sns.boxplot(x='Groups', y='IQ score', data=data)
# Despine
sns.despine(left=True, right=True, top=True)
# Add title
plt.title('IQ scores for different test groups')
# Show plot
plt.show()
```
