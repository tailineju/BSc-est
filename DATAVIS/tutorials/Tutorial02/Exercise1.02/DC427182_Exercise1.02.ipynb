## Slicing

```python
# 01: mean of the intersection of the second and third rows and the second and third columns
print(dataset[1:3, 1:3])
np.mean(dataset[1:3, 1:3])
```

```python
# 02: mean of the subset of every two elements from the 5th row
print(dataset[4, ::2])
np.mean(dataset[4,::2])
```

```python
# 03: mean of the of the last row
np.mean(dataset[-1,::1])
```

## Splitting

```python
# 01: horizontal split of the dataset into 3 equal parts (arrays)
h_split = np.hsplit(dataset,(3))
np.shape(h_split)
```

```python
# 02: vertical split of the first array into 2 equal parts (arrays)
v_split = np.vsplit(h_split[0],2)
np.shape(v_split)
```

```python
# 03: comparing the shapes
print("Original dataset:",dataset.shape)
print("First array of the vertical split:", v_split[0].shape)
```

## Iterating

```python
# 01: one value per row followed by its index (row number)
i = 0
for x in np.nditer(dataset):
    print(x, i)
    i += 1
```

```python
# 02: one value per row preceded by its index (column, row)   
for i, value in np.ndenumerate(dataset):
    print(i, value)
```
