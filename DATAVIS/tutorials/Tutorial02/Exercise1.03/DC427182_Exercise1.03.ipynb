## Combining

```python
# 11:  splitting horizontally and vertically 
thirds = np.hsplit(dataset, (3)) #spliting in three parts horizontally (3 arrays)
halfed_first = np.vsplit(thirds[0], (2)) #splitting the first array in 2 parts vertically
halfed_first[0]
```

```python
# 12: recombining the vertical split
first_col = np.vstack([halfed_first[0], halfed_first[1]])
first_col
```

```python
# 13: recombining the horizontal split
first_second_col = np.hstack([first_col, thirds[1]])
first_second_col
```

```python
# 14: recombining to the original state
full_data = np.hstack([first_second_col, thirds[2]])
full_data
```
