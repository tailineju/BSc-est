```python
#02.04 In the dataset, each sample has a certain weight that reflects the weight for the total dataset. Therefore, we cannot simply calculate the median. Use the given weighted_median function in the following code to compute the median:

# Function for a 'weighted' median
def weighted_frequency(values, weights):
    weighted_values = []
    for value, weight in zip(values, weights):
        weighted_values.extend(np.repeat(value, weight))
    return weighted_values
def weighted_median(values, weights):
    return np.median(weighted_frequency(values, weights))
```

```python
#03.01: Load the Anscombe data set from the Seaborn library, and print it out, using the starter coding provided.
#03.02: Import the package matplotlib with all of the plotting utilities found under a subpackage called pyplot, and give it an abbreviated name plt.
#03.03: Create four subsets of data from the dataset anscombe, as shown in the starter coding above.
#03.04: Verify that each of the four subsets of data from anscombe has the same mean, variance, and correlation, using the describe() and corr() methods. Print the results for each of the four subsets of data.
#03.05: Plot individual line graphs for each of the four subsets of data from 03.03, using plt.plot, by taking in two vectors of values (x-values, y-values).
#03.06: Re-plot individual graphs using points, for each of the four subsets of data from 03.03, using the plt.plot with the 'o' parameter passed.
#03.07: Plot the graphs for each of the four subsets of data from 03.03, in a single figure, instead of completely separate (overlapped) ones, using matplotlib facilities to create subplots.

def answer_FinalExam_Paper03_Question_03_data():
    # 03.01
    print('03.01')
    import seaborn as sns
    anscombe = sns.load_dataset("anscombe")
    print(anscombe)
    # 03.02
    import matplotlib.pyplot as plt
    # 03.03
    dataset_1 = anscombe[anscombe['dataset'] == 'I']
    dataset_2 = anscombe[anscombe['dataset'] == 'II']
    dataset_3 = anscombe[anscombe['dataset'] == 'III']
    dataset_4 = anscombe[anscombe['dataset'] == 'IV']
    # 03.04
    print('03.04')
    print('Dataset 1\n', dataset_1.describe(), '\n', dataset_1.corr(numeric_only=True))
    print('Dataset 2\n', dataset_2.describe(), '\n', dataset_2.corr(numeric_only=True))
    print('Dataset 3\n', dataset_3.describe(), '\n', dataset_3.corr(numeric_only=True))
    print('Dataset 4\n', dataset_4.describe(), '\n', dataset_4.corr(numeric_only=True))
    # 03.05
    print('03.05')
    plt.plot(dataset_1['x'], dataset_1['y'])
    plt.title('Dataset 1')
    plt.show()
    plt.plot(dataset_2['x'], dataset_2['y'])
    plt.title('Dataset 2')
    plt.show()
    plt.plot(dataset_3['x'], dataset_3['y'])
    plt.title('Dataset 3')
    plt.show()
    plt.plot(dataset_4['x'], dataset_4['y'])
    plt.title('Dataset 4')
    plt.show()
    # 03.06 / 03.07
    print('03.06 / 03.07')
    fig, axs = plt.subplots(2, 2, figsize=(10, 8))
    axs[0, 0].plot(dataset_1['x'], dataset_1['y'], 'o')
    axs[0, 0].set_title('Dataset 1')
    axs[0, 1].plot(dataset_2['x'], dataset_2['y'], 'o')
    axs[0, 1].set_title('Dataset 2')
    axs[1, 0].plot(dataset_3['x'], dataset_3['y'], 'o')
    axs[1, 0].set_title('Dataset 3')
    axs[1, 1].plot(dataset_4['x'], dataset_4['y'], 'o')
    axs[1, 1].set_title('Dataset 4')
    plt.title('Anscombe Data')
    plt.tight_layout()
    plt.show()

answer_FinalExam_Paper03_Question_03()
```
