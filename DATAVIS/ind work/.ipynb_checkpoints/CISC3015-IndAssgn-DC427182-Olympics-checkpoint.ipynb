```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1) reading datasets
athletes = pd.read_csv('C:/Users/User/Documents/GitHub/gradest-1/DATAVIS/ind work/datasets/athlete_events.csv')
regions = pd.read_csv('C:/Users/User/Documents/GitHub/gradest-1/DATAVIS/ind work/datasets/noc_regions.csv')
```

```python
# 2) print the first 5 rows of the first dataset
athletes.head()
```

```python
# 3) print the first 5 rows of the second dataset
regions.head()
```

```python
# 4) join the two dataframes
athletes_df = athletes.merge(regions, how='left', on='NOC')
athletes_df.head()
```

```python
# 5) print the shape of the merged dataframe
athletes_df.shape
```

```python
# 6)  rename the last two columns to make column names consistent
athletes_df.rename(columns={'region':'Region', 'notes':'Notes'}, inplace=True);
athletes_df.head()
```

```python
# 7) get the basic information of the merged dataframe
athletes_df.info()
```

```python
# 8) display some statistics of the merged dataframe
athletes_df.describe()
```

```python
# 9) check null values from the merged dataframe
nan_values = athletes_df.isna()
nan_columns = nan_values.any()
nan_columns
```

```python
# 14) get the top ten countries participating in Olympics since 1896
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
# 15) plot the top 10 countries
plt.figure(figsize=(12,6))
plt.title('Overall Participation by Country')
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette='Set2');
```

```python
# 16) get the age distribution of the participants (athletes)
plt.figure(figsize=(12, 6))
plt.title('Age Distribution of the Athletes')
plt.xlabel('Age')
plt.ylabel('Number of Participants')
plt.hist(athletes_df.Age, bins=np.arange(10, 80, 2), color='orange', edgecolor='white');
```

```python
# 19) get Male and Female Participants
gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
# 20)  get pie plot for male and female athletes
plt.figure(figsize=(12, 6))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=150, shadow=True)
```

```python
# 22) get the total number of female athletes in each olympic
female_participants = athletes_df[(athletes_df.Sex=='F')&(athletes_df.Season=='Summer')][['Sex', 'Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.head()
```

```python
# 23) get the total number of female athletes in each olympic
female_participants = athletes_df[(athletes_df.Sex=='F')&(athletes_df.Season=='Summer')][['Sex', 'Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
# 24) alternative method to achieve the above
womenOlympics = athletes_df[(athletes_df.Sex == 'F')&(athletes_df.Season == 'Summer')]
womenOlympics.head()
```

```python
# 25) plot the distribution of women participation - bar graph
sns.set(style='darkgrid')
plt.figure(figsize=(20, 10))
sns.countplot(x='Year', data=womenOlympics, palette='Spectral')
plt.title('Women Olympic Participation')
```

```python
# 26) plot the distribution of women participation - line graph
part = womenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20, 10))
part.loc[:, 'F'].plot()
plt.title('Olot of Female Athletes over time')
```

```python
# 27)  get athletes with gold medals
gold_medalists = athletes_df[(athletes_df.Medal == 'Gold')]
gold_medalists.head()
```

```python
# 28) take only the values that are different from NaN
gold_medalists = gold_medalists[np.isfinite(gold_medalists['Age'])]
gold_medalists
```

```python
# 29) get gold medalists count beyond 60
gold_medalists['ID'][gold_medalists['Age']>60].count()
```

```python
# 30)  get sporting events for those gold medalists beyond 60
sporting_events = gold_medalists['Sport'][gold_medalists['Age']>60]
```

```python
# 31) plot the sporting events for gold medalists beyond 60
plt.figure(figsize=(10, 5))
plt.tight_layout()
sns.countplot(sporting_events)
plt.title('Gold Medals for Athletes over 60 Years')
```

```python
# 32)  get count of gold medals from each participating country in the Olympic Games
gold_medalists.Region.value_counts().reset_index(name='Medal').head(5)
```

```python
# 33)  plot the gold medalists from each participating country
totalGoldMedals = gold_medalists.Region.value_counts().reset_index(name='Medal').head(6)
g = sns.catplot(x='Region', y='Medal', data=totalGoldMedals,
height=5, kind='bar', palette='rocket')
g.despine(left=True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of Medals")
plt.title("Gold Medals per Country")
```

```python
# 34) analyze 2016 Rio Summer Olympics
max_year =athletes_df.Year.max()
print(max_year)
team_names = athletes_df[(athletes_df.Year == max_year)
& (athletes_df.Medal == 'Gold')].Team
team_names.value_counts().head(10)
```

```python
# 35) plot the distribution of the gold medalists in 2016 Rio Summer Olympics
sns.barplot(x=team_names.value_counts().head(20),
y=team_names.value_counts().head(20).index)
plt.ylabel(None)
plt.xlabel('Countrywise Medals for the Year 2016')
```

```python
# 36) get scatterplot to visualize the height vs weight distribution
# of medal-winning athletes from each country in the Olympics
not_null_medals = athletes_df[(athletes_df['Height'].notnull())
& (athletes_df['Weight'].notnull())]
plt.figure(figsize=(12, 10))
axis = sns.scatterplot(x='Height', y='Weight', data=not_null_medals, hue='Sex')
plt.title('Height vs Weight of Olympic Medalists')
```

```python
# 37) Only two people have ever won gold medals in both the Summer and the Winter Olympics.
# Who are these two persons? Which country or countries are they from?
gold_medalists = athletes_df[(athletes_df.Medal == 'Gold')]

gold_ws = gold_medalists.groupby('ID').filter(lambda x: len(x) > 1 and x['Season'].nunique() > 1)
gold_ws[['ID', 'Name', 'Team', 'Season']].drop_duplicates()
```
